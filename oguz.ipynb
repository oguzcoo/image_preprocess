```python
def do_all_algorithms():
    global gaussianBlurKernel
    global gaussianBlurSigmaX
    global thresholdType
    global medianBlurKernelSize
    global backgroundKernelSize
    global weightAlpha
    global weightBeta
    global weightGamma
    global claheClipLimit
    global claheTileGridSize
    global resolutionX
    global resolutionY


    image = cv2.imread(filename)
    # create file dialog to select image file from ipywidget file dialog
    # image = cv2.imread('10005_left.jpg')
    
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    originalImage.value = cv2.imencode('.jpg', image)[1].tobytes()

    image = cv2.resize(image, (resolutionX, resolutionX))
    resizedImage.value = cv2.imencode('.jpg', image)[1].tobytes()

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    

    rgbImageImage.value = cv2.imencode('.jpg', image)[1].tobytes()

    copy = image.copy()
    copy = cv2.cvtColor(copy, cv2.COLOR_RGB2GRAY)
    rgb2grayImage.value = cv2.imencode('.jpg', copy)[1].tobytes()

    blur = cv2.GaussianBlur(copy, (gaussianBlurKernel, gaussianBlurKernel), gaussianBlurSigmaX)
    gaussianBlurImage.value = cv2.imencode('.jpg', blur)[1].tobytes()

    thresh = cv2.threshold(blur, 10, 255, thresholdType)[1]
    thresholdImage.value = cv2.imencode('.jpg', thresh)[1].tobytes()

    # CONTOUR DETECTION
    contour = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contour = contour[0][0]
    contour = contour[:, 0, :]

    # GET COORDINATES
    x1 = tuple(contour[contour[:, 0].argmin()])[0]
    y1 = tuple(contour[contour[:, 1].argmin()])[1]
    x2 = tuple(contour[contour[:, 0].argmax()])[0]
    y2 = tuple(contour[contour[:, 1].argmax()])[1]

    # fill contour image to image widget value
    contourImage.value = cv2.imencode('.jpg', contour)[1].tobytes()


    #Crop Images Again to Destroy Black Area
    x = int(x2 - x1) * 4 // 50
    y = int(y2 - y1) * 4 // 50

    # THRES FOR CROPPED IMAGES
    copy2 = image.copy()
    if x2 - x1 > 100 and y2 - y1 > 100:
        copy2 = copy2[y1 + y: y2 - y, x1 + x: x2 - x]
        copy2 = cv2.resize(copy2, (resolutionX, resolutionX))

    # LAB
    lab = cv2.cvtColor(copy2, cv2.COLOR_RGB2LAB)
    rgb2LabImage.value = cv2.imencode('.jpg', lab)[1].tobytes()

    l, a, b = cv2.split(lab)

    # CLAHE - Contrast-Limited Adaptive Histogram Equalization
    clahe = cv2.createCLAHE(clipLimit=claheClipLimit, tileGridSize=((claheTileGridSize, claheTileGridSize)))
    # claheImage.value = cv2.imencode('.jpg', clahe)[1].tobytes()

    # plt.hist(l.flatten, 25, [0,256], color="b")
    # plt.show()
    cl = clahe.apply(l)
    
   
    # MERGING LAB
    merge = cv2.merge((cl, a, b))
    mergedImage.value = cv2.imencode('.jpg', cl)[1].tobytes()

    final_img = cv2.cvtColor(merge, cv2.COLOR_LAB2RGB)
    lab2rgbImage.value = cv2.imencode('.jpg', final_img)[1].tobytes()

    med_blur = cv2.medianBlur(final_img, medianBlurKernelSize)
    medianBlurImage.value = cv2.imencode('.jpg', med_blur)[1].tobytes()


    back_ground = cv2.medianBlur(final_img, backgroundKernelSize)
    backgroundImage.value = cv2.imencode('.jpg', back_ground)[1].tobytes()

    # MASK FOR BLEEDING VEIN
    mask = cv2.addWeighted(med_blur, weightAlpha, back_ground, weightBeta, weightGamma)
    maskedImage.value = cv2.imencode('.jpg', mask)[1].tobytes()


    weightImage.value = cv2.imencode('.jpg', mask)[1].tobytes()

    final = cv2.bitwise_and(mask, med_blur)
    bitwiseImage.value = cv2.imencode('.jpg', final)[1].tobytes()
```

```python
def handle_change(change):
    global gaussianBlurKernel
    global gaussianBlurSigmaX
    global medianBlurKernelSize
    global backgroundKernelSize
    global weightAlpha
    global weightBeta
    global weightGamma
    global claheClipLimit
    global claheTileGridSize
    global thresholdType
    global resolutionX
    global resolutionY
    global i
    global filename
    
    gaussianBlurKernel = sliderGaussianBlurKernel.value
    gaussianBlurSigmaX = sigmaxslider.value
    medianBlurKernelSize = medianBlurKernelSizeSlider.value
    backgroundKernelSize = backgroundKernelSizeslider.value
    weightAlpha = weightAlphaSlider.value
    weightBeta = weightBetaSlider.value
    weightGamma = weightGammaSlider.value
    claheClipLimit = claheClipLimitSlider.value
    claheTileGridSize = clahetileGridSizeSlider.value
    thresholdType = thresh_constants[threshold_dropdown.value]
    filename = file_dropdown.value
    

    resolutionX = sliderResolutionX.value
    resolutionY = sliderResolutionY.value
    do_all_algorithms()
    
    i = i + 1
    label.value = "Changed " + str(i) + " times"

sliderGaussianBlurKernel.observe(handle_change, names='value')
sigmaxslider.observe(handle_change, names='value')
medianBlurKernelSizeSlider.observe(handle_change, names='value')
backgroundKernelSizeslider.observe(handle_change, names='value')
weightAlphaSlider.observe(handle_change, names='value')
weightBetaSlider.observe(handle_change, names='value')
weightGammaSlider.observe(handle_change, names='value')
claheClipLimitSlider.observe(handle_change, names='value')
clahetileGridSizeSlider.observe(handle_change, names='value')
threshold_dropdown.observe(handle_change, names='value')
sliderResolutionX.observe(handle_change, names='value')
sliderResolutionY.observe(handle_change, names='value')
file_dropdown.observe(handle_change, names='value')

def back_to_default_button_clicked(b):
    global gaussianBlurKernel
    global gaussianBlurSigmaX
    global medianBlurKernelSize
    global backgroundKernelSize
    global weightAlpha
    global weightBeta
    global weightGamma
    global claheClipLimit
    global claheTileGridSize
    global thresholdType
    global resolutionX
    global resolutionY
    global i
    global filename
    
    gaussianBlurKernel = defaultGaussianBlurKernel
    gaussianBlurSigmaX = defaultGaussianBlurSigmaX
    medianBlurKernelSize = defaultMedianBlurKernelSize
    backgroundKernelSize = defaultBackgroundKernelSize
    weightAlpha = defaultWeightAlpha
    weightBeta = defaultWeightBeta
    weightGamma = defaultWeightGamma
    claheClipLimit = defaultClaheClipLimit
    claheTileGridSize = defaultClaheTileGridSize
    thresholdType = cv2.THRESH_BINARY
    resolutionX = defaultResolutionX
    resolutionY = defaultResolutionY
    # filename = jpg_files[0]
    
    sliderGaussianBlurKernel.value = gaussianBlurKernel
    sigmaxslider.value = gaussianBlurSigmaX
    medianBlurKernelSizeSlider.value = medianBlurKernelSize
    backgroundKernelSizeslider.value = backgroundKernelSize
    weightAlphaSlider.value = weightAlpha
    weightBetaSlider.value = weightBeta
    weightGammaSlider.value = weightGamma
    claheClipLimitSlider.value = claheClipLimit
    clahetileGridSizeSlider.value = claheTileGridSize
    threshold_dropdown.value = "Binary Thresholding"
    sliderResolutionX.value = resolutionX
    sliderResolutionY.value = resolutionY
    file_dropdown.value = filename
    
    do_all_algorithms()
    


back_to_default_button.on_click(back_to_default_button_clicked)
# display(grid)
do_all_algorithms()

```
